## Ejercicio 1.1 — Vectores base y ortogonalidad
Los estados |0⟩ y |1⟩ son los vectores base de la computación cuántica.

```python
# Los vectores base
ket_0 = np.array([1, 0], dtype=complex)
ket_1 = np.array([0, 1], dtype=complex)

# Producto interno: ⟨u|v⟩ = suma de u*_i × v_i
inner_00 = np.dot(ket_0.conj(), ket_0)
inner_01 = np.dot(ket_0.conj(), ket_1)
inner_11 = np.dot(ket_1.conj(), ket_1)

print(f'⟨0|0⟩ = {inner_00}  → debe ser 1 (norma = 1)')
print(f'⟨0|1⟩ = {inner_01}  → debe ser 0 (ortogonales)')
print(f'⟨1|1⟩ = {inner_11}  → debe ser 1 (norma = 1)')
print()
print('|0⟩ y |1⟩ son ortonormales: forman una base válida ✓')
```

## Ejercicio 1.2 — Estado en superposición
Un estado cuántico general: |ψ⟩ = α|0⟩ + β|1⟩ donde |α|² + |β|² = 1

```python
alpha = 1/np.sqrt(2)  # amplitud real (caso especial)
beta  = 1/np.sqrt(2)

psi = alpha * ket_0 + beta * ket_1
norma = np.linalg.norm(psi)

print(f'|ψ⟩ = {psi}')
print(f'Norma = {norma:.10f}  (debe ser exactamente 1.0)')
print()
print(f'P(medir |0⟩) = |α|² = {abs(alpha)**2:.4f}')
print(f'P(medir |1⟩) = |β|² = {abs(beta)**2:.4f}')
print(f'Suma = {abs(alpha)**2 + abs(beta)**2:.10f}  (debe ser 1.0)')
print()
print('REGLA CLAVE: Probabilidad = |amplitud|², NO la amplitud directamente')
```

## Ejercicio 1.3 — Amplitudes complejas
Las amplitudes pueden ser números complejos. Las probabilidades siguen siendo reales.

```python
alpha_c = (1 + 1j) / 2   # número complejo
beta_c  = (1 - 1j) / 2

psi_c = alpha_c * ket_0 + beta_c * ket_1

print(f'|ψ_c⟩ = {psi_c}')
print(f'Norma = {np.linalg.norm(psi_c):.10f}')
print()
print(f'P(|0⟩) = |{alpha_c:.2f}|² = {abs(alpha_c)**2:.4f}')
print(f'P(|1⟩) = |{beta_c:.2f}|² = {abs(beta_c)**2:.4f}')
print()
print('Las amplitudes son complejas pero las probabilidades son reales.')
print('La fase compleja importará cuando haya interferencia (semana 4).')
```

## Ejercicio 1.4 — Matrices unitarias
Todas las puertas cuánticas son matrices unitarias: U†U = I

```python
# La puerta Hadamard (la más importante de todas)
H = np.array([[1, 1],
              [1, -1]], dtype=complex) / np.sqrt(2)

H_dag = H.conj().T  # Conjugado traspuesto = dagger (†)
                     # IMPORTANTE: H.T ≠ H†  (solo el conjugado traspuesto es el dagger)

producto = H_dag @ H
print('H†H =')
print(np.round(producto, 10))
print()

es_identidad = np.allclose(producto, np.eye(2))
print(f'¿H†H = I? → {es_identidad}')
print('Si es True: H es unitaria → válida como puerta cuántica ✓')
```

## Función de validación de estado

```python
def es_valido_estado(v: np.ndarray, tol: float = 1e-10) -> bool:
    """Devuelve True si v es un estado cuántico válido (norma = 1)."""
    return bool(abs(np.linalg.norm(v) - 1.0) < tol)

# Pruebas
estados_prueba = [
    ('|0⟩',                    np.array([1, 0], dtype=complex)),
    ('|+⟩ = (|0⟩+|1⟩)/√2',    np.array([1, 1], dtype=complex) / np.sqrt(2)),
    ('Sin normalizar [1, 1]',   np.array([1, 1], dtype=complex)),
    ('Sin normalizar [0.5, 0]', np.array([0.5, 0], dtype=complex)),
    ('Complejo válido',         np.array([(1+1j)/2, (1-1j)/2], dtype=complex)),
]

for nombre, v in estados_prueba:
    valido = es_valido_estado(v)
    icono = '✅' if valido else '❌'
    print(f'{icono} {nombre}: norma={np.linalg.norm(v):.4f} → válido={valido}')
```
